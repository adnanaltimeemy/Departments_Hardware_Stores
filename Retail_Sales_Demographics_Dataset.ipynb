```python
# Step 2: Import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display settings
pd.set_option('display.max_columns', None)

print("Libraries loaded successfully")
```

```python
# Step 3: Define dataset structure

columns = [
    "Company",
    "Category",
    "Year",
    "Revenue (£M)",
    "Profit Margin (%)",
    "Profit (£M)",
    "Age Group",
    "Gender",
    "Income Level",
    "Region"
]

print("Dataset columns created")
```

```python
# Step 4: Debenhams dataset

debenhams_data = [
    ["Debenhams", "Fashion", 2024, 320, 15, 48, "25-44", "Female 65%", "Middle income", "UK"],
    ["Debenhams", "Beauty", 2024, 140, 18, 25.2, "18-34", "Female 75%", "Middle income", "UK"],
    ["Debenhams", "Home", 2024, 98, 12, 11.76, "30-55", "Female 60%", "Middle income", "UK"]
]

df_debenhams = pd.DataFrame(debenhams_data, columns=columns)

df_debenhams
```

```python
# Step 5: Primark dataset

primark_data = [
    ["Primark", "Fashion", 2024, 9448, 11.7, 1105, "16-35", "Female 60%", "Low-Middle income", "Global"],
    ["Primark", "Beauty", 2024, 850, 10.5, 89.25, "18-30", "Female 70%", "Low-Middle income", "Global"],
    ["Primark", "Home", 2024, 620, 9.5, 58.9, "25-44", "Female 55%", "Low-Middle income", "Global"]
]

df_primark = pd.DataFrame(primark_data, columns=columns)

df_primark
```

```python
# Step 6: Combine datasets

df = pd.concat([df_debenhams, df_primark], ignore_index=True)

df
```

```python
# Step 7: Save dataset

df.to_csv("retail_sales_profit_demographics.csv", index=False)

print("Dataset saved successfully")
```

```python
# Step 8: Basic dataset info

df.info()

df.describe()
```

```python
# Step 9: Revenue by company

revenue_summary = df.groupby("Company")["Revenue (£M)"].sum()

revenue_summary
```

```python
# Step 10: Visualization

plt.figure(figsize=(8,5))

sns.barplot(
    data=df,
    x="Company",
    y="Revenue (£M)",
    estimator=sum
)

plt.title("Total Revenue by Company")
plt.ylabel("Revenue (£M)")
plt.show()
```
