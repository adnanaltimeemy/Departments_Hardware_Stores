```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Make plots look nice
sns.set(style="whitegrid")
```

```python
# Create dataset dictionary

data = {
    "Company": [
        "Dunnes Stores",
        "Lidl Stiftung & Co. KG"
    ],
    
    # Revenue in billions EUR
    "Revenue_Billion_EUR": [
        6.5,      # estimated Dunnes
        132.1     # actual Lidl global revenue
    ],
    
    # Profit Margin (%)
    "Profit_Margin_Percent": [
        10.6,     # Dunnes NI estimate
        2.3       # Lidl typical retail margin
    ],
    
    # Market share (% estimate)
    "Market_Share_Percent": [
        25.0,    # Dunnes Ireland approx
        14.2     # Lidl Ireland approx
    ],
    
    # Number of Stores
    "Stores": [
        138,
        12200
    ]
}

# Convert to DataFrame
df_sales = pd.DataFrame(data)

# Calculate Profit
df_sales["Profit_Billion_EUR"] = (
    df_sales["Revenue_Billion_EUR"] * df_sales["Profit_Margin_Percent"] / 100
)

df_sales
```

```python
demographics = {
    "Age_Group": [
        "18-25",
        "26-35",
        "36-50",
        "51-65",
        "65+"
    ],
    
    "Dunnes_Customers_Percent": [
        15,
        22,
        30,
        20,
        13
    ],
    
    "Lidl_Customers_Percent": [
        18,
        25,
        28,
        18,
        11
    ],
    
    "Average_Income_EUR": [
        22000,
        32000,
        45000,
        40000,
        30000
    ]
}

df_demo = pd.DataFrame(demographics)

df_demo
```

```python
# Merge datasets example

combined = {
    "Company": ["Dunnes Stores", "Lidl Stiftung"],
    
    "Revenue_Billion": df_sales["Revenue_Billion_EUR"],
    "Profit_Billion": df_sales["Profit_Billion_EUR"],
    "Profit_Margin": df_sales["Profit_Margin_Percent"]
}

df_combined = pd.DataFrame(combined)

df_combined
```

```python
df_sales.describe()
```

```python
plt.figure()

plt.bar(df_sales["Company"], df_sales["Revenue_Billion_EUR"])

plt.title("Revenue Comparison")
plt.xlabel("Company")
plt.ylabel("Revenue (Billion EUR)")

plt.show()
```

```python
plt.figure()

plt.bar(df_sales["Company"], df_sales["Profit_Margin_Percent"])

plt.title("Profit Margin Comparison")
plt.xlabel("Company")
plt.ylabel("Profit Margin (%)")

plt.show()
```

```python
df_demo.plot(
    x="Age_Group",
    y=["Dunnes_Customers_Percent", "Lidl_Customers_Percent"],
    kind="bar"
)

plt.title("Customer Age Demographics")
plt.ylabel("Customer Percentage")

plt.show()
```

```python
df_sales.to_csv("sales_dataset.csv", index=False)
df_demo.to_csv("demographics_dataset.csv", index=False)

print("Datasets saved successfully.")
```
